```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Define column names
# col_names = ["PercentageFragments", "NumberFragments", "FragmentsAssigned", "RankCode", "NCBITaxID", "ScientificName"]

# Define filepath
filepath = "/mnt/archgen/users/schumacher/bachelorthesis/05-results/KRK2_ar_bac_plas_vir_hum_fun_plant_uni_vert.tsv"

# Load the data
data = pd.read_csv(filepath, sep="\t")  # , names=col_names
genera_df = data[data['rank'] == 'genus']

genera_df
# data
```

```python
# Add up values for each genus and create column with summed values
genera_df['total'] = genera_df.iloc[:, 3:-1].sum(axis=1)
genera_df
```

```python
# Sort by column 'total' descending
sorted_genera_df = genera_df.sort_values(by='total', ascending=False)
sorted_genera_df
```

```python
# Keep top 20 genera and sum up the rest as "other"
# Split the DataFrame into top 20 and the rest
top_20 = sorted_genera_df.iloc[:15]
rest = sorted_genera_df.iloc[15:]

# Sum up the values of the rest of the rows
sum_values = rest.sum(numeric_only=True)
sum_values['taxonomy_id'] = 'other'
sum_values['name'] = 'other'
sum_values['rank'] = 'genus'

# Append the new row to the top 20 rows
result = pd.concat([top_20, pd.DataFrame([sum_values])], ignore_index=True)

result
```

```python
for samp_col in result.columns:
    if samp_col.startswith('BGG'):
        sum = result[samp_col].sum()
        result[samp_col] = result[samp_col].apply(lambda x: x / sum * 100)

result
```

```python
# Mapping dictionary
sample_name_mapping = {
    "BGG_BOSW": "0_NA_B",
    "BGG_POSDC": "1_11_P",
    "BGG_KOSDC": "1_16_K",
    "BGG_LOSDA": "2_12_L",
    "BGG_NOSDA": "2_16_N",
    "BGG_ROSDA": "2_19_R",
    "BGG_QOSDA": "3_06_Q",
    "BGG_MOSDA": "3_16_M",
    "BGG_OOSDC": "3_19_O"
}
 
# Rename columns that start with 'BGG'
new_columns = {col: sample_name_mapping.get(col, col) for col in result.columns}
result.rename(columns=new_columns, inplace=True)

# rearrange columns
result = result[['taxonomy_id', 'name', 'rank', '0_NA_B', '1_11_P', '1_16_K', '2_12_L', '2_16_N', '2_19_R', '3_06_Q', '3_16_M', '3_19_O', 'total']]

result
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Set the style of seaborn for better visualization
sns.set_style("whitegrid")

# Increase font size of all elements
sns.set(font_scale=1.5)

# Get 21 distinct colors
colors = plt.cm.tab20.colors  # This gives 20 colors
colors = list(colors)
colors.append((0.5, 0.2, 0.8, 1))  # Adding a 21st color

# Prepare data for plotting
samples = result.columns[3:-1]
labels = result['name'].values

# Plotting
fig, ax = plt.subplots(figsize=(15, 10))
bottom = [0] * len(samples)

for index, row in result.iterrows():
    ax.bar(samples, row[samples], label=row['name'], bottom=bottom, color=colors[index])
    bottom += row[samples].values

ax.set_ylabel('Relative abundance [%]')
ax.set_xlabel('Sample name')

# Create the legend and set the text to italic
legend = ax.legend(loc='upper left', bbox_to_anchor=(1, 1))
legend.get_frame().set_facecolor('white')
for text in legend.get_texts():
    if text.get_text() != 'other':
        text.set_style('italic')

plt.xticks(rotation=45)
plt.tight_layout()

# Set the face color of the plot
ax.set_facecolor('white')

# Set the color of the axes to black
for spine in ax.spines.values():
    spine.set_edgecolor('black')

# Save the figure with a white background
plt.savefig('top_genera.jpg', dpi=600, bbox_inches='tight', facecolor='white')
```
